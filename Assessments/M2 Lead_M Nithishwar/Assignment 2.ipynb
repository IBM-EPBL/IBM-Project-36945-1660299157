# **Finding and Handling the outliers**

```python
sns.boxplot(x=ds['Age'])
```

```python
ds['Age'].hist()
```

```python
print('skewness value of Age: ',ds['Age'].skew())
```

```python
Q1 = ds['Age'].quantile(0.25)
Q2 = ds['Age'].quantile(0.75)
IQR = Q2 - Q1
whisker_width = 1.5
lower_whisker = Q1 -(whisker_width*IQR)
upper_whisker = Q2 +(whisker_width*IQR)
ds['Age']=np.where(ds['Age']>upper_whisker,upper_whisker,np.where(ds['Age']<lower_whisker,lower_whisker,ds['Age']))
sns.boxplot(x=ds['Age'])
```

```python
Standardisation = preprocessing.StandardScaler()
x_after_Standardisation = Standardisation.fit_transform(X)
print ("\nAfter Standardisation : \n", x_after_Standardisation)
```
